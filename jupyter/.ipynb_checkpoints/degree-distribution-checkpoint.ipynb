# Degree distribution

```python
import os
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
data_dir = "../data/"
os.listdir(data_dir)
```

```python
filename = data_dir + "degree-distribution.txt"
```

```python
k, N_k = np.loadtxt(filename, unpack=True)
```

```python
plt.plot(k, N_k)
```

```python
x, y = np.log(k), np.log(N_k)
plt.plot(x, y)
```

```python
stats.linregress(x, y)
```

```python
x, y = np.log(k), np.log(N_k)
plt.plot(x, y)


idx = (x > 2) & (x < 4.5)

plt.plot(x[idx], y[idx])


stats.linregress(x[idx], y[idx])
```
