# Degree distribution

```python
import os
import numpy as np
import matplotlib.pyplot as plt # plotting
import scipy.stats as stats # for linear regression
import glob
```

```python
dir_data = '../data/'
os.listdir(dir_data)
```

```python
filename = dir_data + "degree-distribution_m3_.txt"
```

```python
print(filename)
```

```python
k, N_k = np.loadtxt(filename, unpack=True)
```

```python
plt.plot(k, N_k)
```

```python
log_k  = np.log(k)
log_N_k = np.log(N_k) 
plt.plot(log_k, log_N_k)
plt.xlabel(r"$\log(k)$")
plt.ylabel(r"$\log(N(k))$")
```

```python

log_k  = np.log(k)
log_N_k = np.log(N_k) 

idx  = (log_k > 2) & (log_k < 4.5)
plt.plot(log_k, log_N_k, label='full')
plt.plot(log_k[idx], log_N_k[idx], label='selected')
plt.xlabel(r"$\log(k)$")
plt.ylabel(r"$\log(N(k))$")
plt.legend()
```

```python
regress = stats.linregress(log_k[idx], log_N_k[idx])
print(regress)
```

```python
files = glob.glob(dir_data + "degree-distribution_m*_.txt")
print(files)
```

```python
for file in files:
    k, N_k = np.loadtxt(file, unpack=True)
    log_k  = np.log(k)
    log_N_k = np.log(N_k) 

    
    plt.plot(log_k, log_N_k, label=file)
    
plt.xlabel(r"$\log(k)$")
plt.ylabel(r"$\log(N(k))$")
plt.legend()
```
